```python
import os
import sys
# sys.path.append(os.path.abspath('../'))
# print(sys.path)
from evaluate import read_gold_result, read_pred_results 
import random
import tabulate
from IPython.display import display_html
```

```python
pred_model_1 = 'schema'
pred_model_2 = 'code'
pred_model_3 = 'sub_args'
pred_model_4 = 'explanation'
split = 'test'

config_file = "../dataset_processing/data_config/event/phee2.yaml"
gold_file = "../data/datasets/phee2/source/%s.json"%split
# pred_file_1 = "hf_models/%s/test_preds_record.txt"%(pred_model_1)
# pred_file_2 = "hf_models/%s/test_preds_record.txt"%(pred_model_2)
pred_folder_1 = "../chatgpt_zero_shot/output/%s/%s"%(pred_model_1, split)
pred_folder_2 = "../chatgpt_zero_shot/output/%s/%s"%(pred_model_2, split)
pred_folder_3 = "../chatgpt_zero_shot/output/%s/%s"%(pred_model_3, split)
pred_folder_4 = "../chatgpt_zero_shot/output/%s/%s"%(pred_model_4, split)

gold_instances = read_gold_result(gold_file)
pred_instances_m1 = read_pred_results(pred_folder_1, gold_instances)
pred_instances_m2 = read_pred_results(pred_folder_2, gold_instances)
pred_instances_m3 = read_pred_results(pred_folder_3, gold_instances)
pred_instances_m4 = read_pred_results(pred_folder_4, gold_instances)
```

```python
while True:
    sample_id = random.randint(0, len(gold_instances)-1)
    

    gold = gold_instances[sample_id]
    pred1 = pred_instances_m1[sample_id]
    pred2 = pred_instances_m2[sample_id]
    pred3 = pred_instances_m3[sample_id]
    pred4 = pred_instances_m4[sample_id]

    context = gold['context']
    instance_id = gold['id'][0]

    keys = set(list(gold.keys())+list(pred1.keys())+list(pred2.keys())+list(pred3.keys())+list(pred4.keys()))
    keys.remove('id')
    keys.remove('is_mult')
    keys.remove('context')
    keys = list(keys)
    keys.sort(reverse=True)


    data_table = []
    data_table.append(["", "gold", pred_model_1, pred_model_2, pred_model_3, pred_model_4])
    for key in keys:
        line = [key]
        if key in gold:
            line.append("|".join(gold[key]))
        else:
            line.append("")
        if key in pred1:
            line.append("|".join(pred1[key]))
        else:
            line.append("")
        if key in pred2:
            line.append("|".join(pred2[key]))
        else:
            line.append("")
        if key in pred3:
            line.append("|".join(pred3[key]))
        else:
            line.append("")
        if key in pred4:
            line.append("|".join(pred4[key]))
        else:
            line.append("")
        data_table.append(line)
    
    query_arg = "Population"
    diff_pred1 = False
    diff_pred2 = False
    diff_pred3 = False
    diff_pred4 = False
    for line in data_table:
        if query_arg in line[0]:
            if line[1] != line[2]:
                diff_pred1 = True
            if line[1] != line[3]:
                diff_pred2 = True

    if len(data_table) > 1 and diff_pred1:
    # if instance_id == '8467620_3':
        table = tabulate.tabulate(data_table, tablefmt='html', maxcolwidths=20)
        print(instance_id, ":", context)
        display_html(table)

        break

```
