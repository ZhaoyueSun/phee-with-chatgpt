```python
from tabulate import tabulate
from collections import OrderedDict
import json
```

```python
global present_args
present_args = ['Arguments', 'MainArgs', 'SubArgs', 'Subject', 'Subject.Age', 'Subject.Disorder', 'Subject.Gender', 'Subject.Population',
    'Subject.Race', 'Treatment', 'Treatment.Drug', 'Treatment.Disorder', 'Treatment.Route', 'Treatment.Dosage',
    'Treatment.Time_elapsed', 'Treatment.Duration', 'Treatment.Freq', 'Combination.Drug', 'Effect']

def read_rst(rst_path):
    rst = OrderedDict({
        'EM_P': OrderedDict(),
        'EM_R': OrderedDict(),
        'EM_F1': OrderedDict(),
        'MICRO_P': OrderedDict(),
        'MICRO_R': OrderedDict(),
        'MICRO_F1': OrderedDict(),
        'INCONSISTENT_ARGS': OrderedDict(),
        'MISS_ARGS': OrderedDict(),
        'RELUCTANT_ARGS': OrderedDict(),

    }
    )
    
    with open(rst_path, 'r') as f:
        raw_rst = json.load(f)
        for m in rst:
            for arg in present_args:
                if 'CLS_%s_%s'%(arg,m) in raw_rst:
                    rst[m][arg] = round(raw_rst['CLS_%s_%s'%(arg,m)], 2)
                else:
                    rst[m][arg] =  0
    return rst
```

```python

rst_data = dict()

models = ["BM25-Type", "TreeKernelW-Type"]
for model_name in models:
    rst_data[model_name] = read_rst(os.path.join("chatgpt_few_shot/output",model_name+"_demo_5",'test_result.json'))
    

table = []
first_line = [""]
for x in models:
    first_line += [x]
    # first_line += [x.split('_')[4]]
    first_line += ["" for _ in range(len(rst_data[x])-1)]

# first_line = ['','llama_30b_bm25', 'flan-t5-xxl_bm25']

table.append(first_line)
second_line = [""]
for k in models:
    for m in rst_data[k]:
        second_line.append(m)
table.append(second_line)

for arg in present_args:
    line = [arg]
    for k in models:
        for m in rst_data[k]:
            line.append(rst_data[k][m][arg])
    table.append(line)

tabulate(table, tablefmt='html', maxcolwidths=5)
```
